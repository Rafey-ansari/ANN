```python
from bs4 import BeautifulSoup 
import pandas as pd
import requests
import re
```

```python
places=[]
placestovisit="https://www.holidify.com/country/india/places-to-visit.html"
response1=requests.get(placestovisit)
soup1=BeautifulSoup(response1.content,"html.parser")

for i in soup1.find_all('h3'):
    a=i.text
    res1 = " ".join(re.findall("[a-zA-Z]+",a))
    if res1=='Leh Ladakh':
        places.append('Ladakh')
    else:
        places.append(res1)
```

```python
list1=[]
for i in places:
    url=f"https://www.holidify.com/places/{i}/"
    response=requests.get(url)
    soup=BeautifulSoup(response.content,"html.parser")
    try:
        city=soup.find('div',{'class':'col-12 col-lg-6 right'}).find('h1').text.strip()
    except Exception:
        city=None
    try:
        rating=soup.find('span',{'class':"rating-badge"}).text.strip()[:-4]
    except Exception:
        rating=None
    try:
        price=soup.find('div',{'class':'col-12 col-lg-6 right'}).find('div',{'class':'mb-2 text-center-mobile'}).text.strip()[:8]
    except Exception:
        price=None
    try:
        places_to_visit=soup.find('div',class_='row no-gutters').find('div',{'class':'col-lg-8 pr-lg-2'}).find_all('div',class_='col-6 col-md-4 ptv-item')
        nearby= [i.text.strip() for i in places_to_visit]
    except Exception:
        neaby=None
    list1.append([city, rating, price, nearby])
    

df=pd.DataFrame(list1,columns=['city','rating','price','nearby'])
df
```

```python
cities=[]
url2='http://www.incredibleindia-tourism.org/north-india-tourism/'
response2=requests.get(url2)
soup2=BeautifulSoup(response2.content,'html.parser')
for i in soup2.find('div', class_='table-responsive').find_all('li'):
    cities.append(i.text)
    
url3='http://www.incredibleindia-tourism.org/south-india-tourism/'
response3=requests.get(url3)
soup3=BeautifulSoup(response3.content,'html.parser')
for i in soup3.find('div',class_='table-responsive').find_all('li'):
    cities.append(i.text)
    
url4='http://www.incredibleindia-tourism.org/north-east-india-tourism/'
response4=requests.get(url4)
soup4=BeautifulSoup(response4.content,'html.parser')
for i in soup4.find('div',class_='table-responsive').find_all('li'):
    cities.append(i.text)
    
url5='http://www.incredibleindia-tourism.org/west-central-india-tourism/'
response5=requests.get(url5)
soup5=BeautifulSoup(response5.content,'html.parser')
for i in soup5.find('div',class_='table-responsive').find_all('li'):
    cities.append(i.text)
    
cities
```

```python
incredible_india=[]
for i in cities:
    
    url1=f'http://www.incredibleindia-tourism.org/weekend-getaways/{i.replace("/","-").lower()}.html'
    response1=requests.get(url1)
    soup1=BeautifulSoup(response1.content,"html.parser")
    travel=soup1.find('div',class_='col-sm-12 col-md-8 col-lg-6 content_area order_one maincontentinner').find_all('p')
    strg=""
    for j in travel[-7:]:
        strg=strg+j.text+" "
    incredible_india.append([i,strg])
incredible_india
    
```

```python
df1=pd.DataFrame(incredible_india,columns=['city','travel description'])
df1
```
